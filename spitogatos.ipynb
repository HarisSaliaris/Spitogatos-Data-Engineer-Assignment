read json files and put the data to dataframe

```python
DataList = []

csvheader = ['Date',
             'Area Id',
             'Category'
             'Listing Type',
             'Living Area Low',
             'Living Area High',
             'Price Low',
             'Price High',
             'New development',
             'Garage',
             'Storage',
             'Balcony',
             'Secure Door',
             'Alarm',
             'Fireplace',
             'Elevator',
             'Garden',
             'Rooms Low',
             'Rooms High',
             'Pets Allowed',
             'Brokers IDS']

dataframe = pd.DataFrame(columns = csvheader)

with open('C:/Users/haris/Desktop/spitogatos_data_engineer_assignment_raw_data/data.json') as f:
    for jsonObj in f:
        DataDict = json.loads(jsonObj)
        DataList.append(DataDict)


for data in DataList:
    #print(data)
    #print("/n")
    #Date 
    if 'date' in data:
        date=data['date']
    else:
        date=''
    #Area Id 
    if 'areaIDs' in data:
        area_id=data['areaIDs']
    else:
        area_id=''
    #Category
    if 'category' in data:
        category=data['category']
    else:
        category=''
    #Listing Type
    if 'listingType' in data:
        listingtype=data['listingType']
    else:
        listingtype=''
    #Living Area Low
    if 'livingAreaLow' in data:
        livingarealow=data['livingAreaLow']
    else:
        livingarealow=''
    #Living Area High
    if 'livingAreaHigh' in data:
        livingareahigh=data['livingAreaHigh']
    else:
        livingareahigh=''
    #Price Low
    if 'priceLow' in data:
        pricelow=data['priceLow']
    else:
        pricelow=''
    #Price High
    if 'priceHigh' in data:
        pricehigh=data['priceHigh']
    else:
        pricehigh=''
    #New development
    if 'newDevelopment' in data:
        newdevelopment=data['newDevelopment']
    else:
        newdevelopment=''
    #Garage
    if 'garage' in data:
        garage=data['garage']
    else:
        garage=''
    #Storage
    if 'storage' in data:
        storage=data['storage']
    else:
        storage=''
    #Balcony
    if 'balcony' in data:
        balcony=data['balcony']
    else:
        balcony=''
    #Secure Door
    if 'secureDoor' in data:
        securedoor=data['secureDoor']
    else:
        securedoor=''
    #Alarm
    if 'alarm' in data:
        alarm=data['alarm']
    else:
        alarm=''
     #Fireplace
    if 'fireplace' in data:
        fireplace=data['fireplace']
    else:
        fireplace=''
    #Elevator
    if 'elevator' in data:
        elevator=data['elevator']
    else:
        elevator=''
    #Garden
    if 'garden' in data:
        garden=data['garden']
    else:
        garden=''
    #Rooms Low
    if 'roomsLow' in data:
        roomslow=data['roomsLow']
    else:
        roomslow=''
    #Rooms High
    if 'roomsHigh' in data:
        roomshigh=data['roomsHigh']
    else:
        roomshigh=''
    #Pets Allowed
    if 'petsAllowed' in data:
        petsallowed=data['petsAllowed']
    else:
        petsallowed=''
    #Brokers IDS
    if 'brokerIDs' in data:
        brokerids=data['brokerIDs']
    else:
        brokerids=''
    
    dataframe=dataframe.append({'Date':date,'Area Id':area_id,'Category':category,'Listing Type':listingtype,'Living Area Low':livingarealow,'Living Area High':livingareahigh,'Price Low':pricelow,'Price High':pricehigh,'New development':newdevelopment,'Garage':garage,'Storage':storage,'Balcony':balcony,'Secure Door':securedoor,'Alarm':alarm,'Fireplace':fireplace,'Elevator':elevator,'Garden':garden,'Rooms Low':roomslow,'Rooms High':roomshigh,'Pets Allowed':petsallowed,'Brokers IDS':brokerids},ignore_index=True)
```
